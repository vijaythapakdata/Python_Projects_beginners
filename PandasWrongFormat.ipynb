```python
import pandas as pd
df=pd.read_csv("/content/dirtydata.csv")
```

```python
print(df.to_string())
```

```python
#in date row you can check row 22 and 26 contains wrong format
#pandas has inbuilt function to_datetime()
df["Date"]=pd.to_datetime(df["Date"])
print(df.to_string())
```

```python
#or you can completetly drop the row
df.drop(["Date"],axis="columns",inplace=True)
```

```python
print(df.to_string())
```

```python
#cleaning wrong data
for x in df.index:
  if df.loc[x,"Duration"]>120:
    df.loc[x,"Duration"]=120
```

```python
print(df.to_string())
```

```python
for i in df.index:
  if df.loc[i,"Pulse"]<100:
    df.loc[i,"Pulse"]=100
```

```python
print(df.to_string())
```
