```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#simple pie chart

list_a=np.array([25,15,25,35])

plt.pie(list_a)
plt.show()
```

```python
#pie chart with lable

y=np.array([35,15,25,25])
mylabels=["Apple","Dates","Cherry","Banana"]

plt.pie(y,labels=mylabels)
plt.show()
```

```python
#exploder pie

y=np.array([35,15,25,25])
mylabels=["Men","Women","Boys","Girls"]
myexploded=[0.2,0,0,0]

plt.pie(y,labels=mylabels,explode=myexploded)
plt.show()
```

```python
#shadow
y=np.array([35,15,25,25])
mylabels=["Potato","Tomato","Onion","Brinjal"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

```python
#pie chart with colors
y=np.array([35,15,25,25])
mylabels=["Potato","Tomato","Onion","Brinjal"]
mycolors=["red","green","yellow","blue"]
myexplode=[0.2,0,0,0]
plt.pie(y,colors=mycolors,labels=mylabels,explode=myexplode,shadow=True)
plt.show()
```

```python
#pie chart with legend
y=np.array([35,15,25,25])

mylabels=["Apple","Mango","Grapes","Dates"]

plt.pie(y,labels=mylabels)
plt.legend()
plt.show()
```

```python
#legend with title
y=np.array([35,15,25,25])

mylabels=["Apple","Mango","Grapes","Dates"]

plt.pie(y,labels=mylabels)
plt.legend(title="Fruits")
plt.show()
```

```python
#pie chart with autopercentage
sports=["Cricekt","Hockey","Football","Tennis","Basketball"]

number_of_pepople=[500,240,300,109,308]

plt.pie(number_of_pepople,labels=sports,autopct="%.1f")
plt.legend(title="People Vs Sports")
plt.show()
```
