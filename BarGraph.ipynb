```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#simple bar graph
x=np.array(["A","B","C","D"])
y=np.array([100,50,230,123])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Graph")
plt.bar(x,y)
plt.show()
```

```python
#bar graph with colors
x=np.array(["A","B","C","D"])
y=np.array([100,50,230,123])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Graph")
plt.bar(x,y,color="red")
plt.show()
```

```python
#bar graph with green color
x=np.array(["A","B","C","D"])
y=np.array([100,50,230,123])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Graph")
plt.bar(x,y,color="green")
plt.show()
```

```python
#horizontal bargraph
x=np.array(["A","B","C","D"])
y=np.array([100,50,230,123])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Graph")
plt.barh(x,y,height=0.1)
plt.show()
```

```python
#bar graph with bar width
x=np.array(["A","B","C","D"])
y=np.array([100,50,230,123])

plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.title("Bar Graph")
plt.bar(x,y, width=0.4)
plt.show()
```
